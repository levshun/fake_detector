## Step 2
Initialize the GeneratedImageDetector class  
models can be downloaded from https://github.com/levshun/fake_detector/releases/download/generated_detection_models_v1/models.zip

```python
detector = GeneratedImageDetector(
    vit_model_path="./models/eva_model.pth",
    convnext_model_path="./models/convnext_model.pth",
    dt_model="./models/final_decisiontree.pkl",
    device="cuda", # can be either "cuda" or "cpu"
)
```

```python
from IPython.display import display, Image as IPImage
display(IPImage(filename="./tests/test_images/bad.jpg", width=300))
```

# Step 3 
Call get_final_score with the path to the image you want to test.  
result['is_fake']: a Boolean indicating whether the image is fake.  
result['prob_of_fake']: a float representing the probability that the image is fake.  



```python
result = detector.get_final_score("./tests/test_images/bad.jpg")
print(f"Predicted label: {result['is_fake']}")
print(f"Probability of being fake: {result['prob_of_fake']}")
```
